## Working on issues identified

```python
# removing spaces and tabs
dataset = dataset.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# converting "?" into missing values
dataset = dataset.replace("?", np.nan)
```

```python
# converting numeric columns
dataset["pcv"] = pd.to_numeric(dataset["pcv"], errors="coerce")
dataset["wc"] = pd.to_numeric(dataset["wc"], errors="coerce")
dataset["rc"] = pd.to_numeric(dataset["rc"], errors="coerce")
```

## Rechecking unique values & data types

```python
# rechecking the unique values
for column in dataset.columns:
    print(column, dataset[column].unique())
```

## Working on missing values

```python
# Importing quan qual function from already created library
from Univariate import Univariate
quan, qual = Univariate.quanQual(dataset)
```

```python
# Replacing missing values in numerical columns (quan) with mean values
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
dataset[quan] = imp.fit_transform(dataset[quan])
```

```python
# Replacing missing values in categorical columns (qual) with mode values
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
dataset[qual] = imp.fit_transform(dataset[qual])
```
